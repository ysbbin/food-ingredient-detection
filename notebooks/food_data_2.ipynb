```python
import os
from collections import defaultdict
import yaml

# ======================================
# âœ… ê²½ë¡œ ì„¤ì •
# ======================================

# í˜„ì¬ .ipynb íŒŒì¼ ê¸°ì¤€ ìƒëŒ€ ê²½ë¡œ
DATA_DIR = "food_data_2"
LABELS_DIR = os.path.join(DATA_DIR, "train", "labels")
YAML_PATH = os.path.join(DATA_DIR, "data.yaml")

# ======================================
# âœ… í´ë˜ìŠ¤ ì´ë¦„ ë¡œë“œ
# ======================================

with open(YAML_PATH, 'r') as f:
    data_yaml = yaml.safe_load(f)

class_names = data_yaml['names']

# ======================================
# âœ… í´ë˜ìŠ¤ë³„ ì´ë¯¸ì§€ ìˆ˜ ì„¸ê¸°
# ======================================

class_image_count = defaultdict(set)

for label_file in os.listdir(LABELS_DIR):
    if label_file.endswith('.txt'):
        label_path = os.path.join(LABELS_DIR, label_file)
        with open(label_path, 'r') as f:
            for line in f:
                class_idx = int(line.split()[0])
                class_image_count[class_idx].add(label_file)

# ======================================
# âœ… ê²°ê³¼ ì¶œë ¥
# ======================================

print(f"ğŸ“‚ [Train] í´ë˜ìŠ¤ë³„ ì´ë¯¸ì§€ ê°œìˆ˜:")

for class_idx in sorted(class_image_count.keys()):
    class_name = class_names[class_idx]
    image_count = len(class_image_count[class_idx])
    print(f"   í´ë˜ìŠ¤ {class_idx} ({class_name}): {image_count}")
```
