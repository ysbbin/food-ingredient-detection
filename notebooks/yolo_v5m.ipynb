```python
import yaml

# ==============================================
# âœ… 1) data.yaml ì ˆëŒ€ ê²½ë¡œ
# ==============================================
yaml_path = r"C:\Users\PC\Desktop\ì¡¸ìž‘\food_data_3\data.yaml"

# ==============================================
# âœ… 2) yolov5 ê¸°ì¤€ ìƒëŒ€ê²½ë¡œë¡œ ìˆ˜ì •í•  ê²½ë¡œ
# ==============================================
new_train = "../food_data_3/train/images"
new_val   = "../food_data_3/valid/images"
new_test  = "../food_data_3/test/images"

# ==============================================
# âœ… 3) YAML ì—´ê¸°
# ==============================================
with open(yaml_path, 'r') as f:
    data = yaml.safe_load(f)

# ==============================================
# âœ… 4) ê²½ë¡œ ìˆ˜ì •
# ==============================================
data['train'] = new_train
data['val']   = new_val
data['test']  = new_test

# ==============================================
# âœ… 5) YAML ë®ì–´ì“°ê¸° ì €ìž¥
# ==============================================
with open(yaml_path, 'w') as f:
    yaml.dump(data, f)

print("âœ… data.yaml ìˆ˜ì • ì™„ë£Œ!")
print(f"train: {data['train']}")
print(f"val:   {data['val']}")
print(f"test:  {data['test']}")
```

```python
import os
import subprocess
from datetime import datetime

# ============================================
# âœ… 1) ê²½ë¡œ ì„¤ì • (ì ˆëŒ€ê²½ë¡œ!)
# ============================================
# âš¡ ë°˜ë“œì‹œ ë³¸ì¸ ê²½ë¡œì— ë§žê²Œ ìˆ˜ì •í•˜ì„¸ìš” âš¡
YOLOV5_DIR = r"C:\Users\PC\Desktop\ì¡¸ìž‘\yolov5"
DATA_YAML = r"C:\Users\PC\Desktop\ì¡¸ìž‘\food_data_3\data.yaml"
WEIGHTS = "yolov5m.pt"

# ============================================
# âœ… 2) í•™ìŠµ íŒŒë¼ë¯¸í„°
# ============================================
EPOCHS = 100
IMG_SIZE = 640
BATCH_SIZE = 16

# runs/train/ì— ê³ ìœ  í´ë”ëª…
EXPERIMENT_NAME = f"exp_food3_v5m_{datetime.now().strftime('%Y%m%d_%H%M%S')}"

# ============================================
# âœ… 3) subprocess ì»¤ë§¨ë“œ
# ============================================
command = [
    "python", "train.py",
    "--img", str(IMG_SIZE),
    "--batch", str(BATCH_SIZE),
    "--epochs", str(EPOCHS),
    "--data", DATA_YAML,
    "--weights", WEIGHTS,
    "--project", "../runs/train",
    "--name", EXPERIMENT_NAME,
    "--device", "0"  # GPU ID
]

print(f"ðŸš€ ì‹¤í–‰ ì»¤ë§¨ë“œ:\n{' '.join(command)}")

# ============================================
# âœ… 4) YOLOv5 í´ë”ë¡œ ì´ë™ (ì¤‘ìš”!)
# ============================================
print("ðŸ“‚ í˜„ìž¬ ìž‘ì—… ë””ë ‰í† ë¦¬:", os.getcwd())
print("ðŸ“‚ YOLOV5_DIR ê²½ë¡œ:", YOLOV5_DIR)

os.chdir(YOLOV5_DIR)
print("ðŸ“‚ ì´ë™ í›„ ìž‘ì—… ë””ë ‰í† ë¦¬:", os.getcwd())

# ============================================
# âœ… 5) subprocessë¡œ ì‹¤ì‹œê°„ ë¡œê·¸ ì¶œë ¥ + ì¸ì½”ë”© ì•ˆì „
# ============================================
process = subprocess.Popen(
    command,
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True,
    encoding="utf-8",
    errors="replace"  # ê¹¨ì§€ëŠ” ë¬¸ìž ëŒ€ì²´
)

for line in process.stdout:
    print(line, end="")

process.wait()
print(f"\nâœ… í•™ìŠµ ì™„ë£Œ! ê²°ê³¼ëŠ” runs/train/{EXPERIMENT_NAME} ì— ì €ìž¥ë¨")
```
