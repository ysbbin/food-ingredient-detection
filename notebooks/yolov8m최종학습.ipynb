```python
from ultralytics import YOLO
import os

# ==========================
# ✅ 절대경로
# ==========================
DATA_YAML_PATH = os.path.abspath("food_data_3/data.yaml")
print("✅ data.yaml 경로:", DATA_YAML_PATH)

# ==========================
# ✅ YOLOv8m 모델 로드
# ==========================
model = YOLO('yolov8m.pt')

# ==========================
# ✅ 학습 결과 저장 위치
# ==========================
PROJECT_PATH = os.path.abspath("runs")
print("✅ 학습 결과 저장 위치:", PROJECT_PATH)

# ==========================
# ✅ 학습 실행 (YOLOv8m)
# ==========================
m
odel.train(
    data=DATA_YAML_PATH,
    epochs=100,
    imgsz=640,
    batch=16,
    project=PROJECT_PATH,
    name='food_data3_exp_v8m',   # 실험 이름 다르게!
    save_period=1
)
```

```python
# 학습 결과 확인 (CSV & 그래프)

import pandas as pd
from IPython.display import Image

# CSV 경로
results_csv = os.path.join(PROJECT_PATH, 'food_data3_exp_v8m', 'results.csv')
df = pd.read_csv(results_csv)

print("✅ 에포크별 검증 지표:")
print(df[['epoch', 'metrics/precision(B)', 'metrics/recall(B)', 'metrics/mAP50(B)']].head())

# 결과 그래프 출력
results_png = os.path.join(PROJECT_PATH, 'food_data3_exp', 'results.png')
Image(filename=results_png, width=800)
```

```python
from ultralytics import YOLO
import os

# 1) 모델 불러오기
model = YOLO("runs/food_data3_exp_v8m/weights/best.pt")

# 2) 테스트 폴더 경로
TEST_FOLDER = os.path.abspath("외부 test")

# 3) 추론 실행
results = model.predict(
    source=TEST_FOLDER,
    conf=0.3,
    save=True,
    project="runs",
    name="predict_external_test_v8m"
)

print("✅ 탐지 결과 저장 폴더:", results[0].save_dir)
```
