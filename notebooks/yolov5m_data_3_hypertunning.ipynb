```python
# YOLOv5~v8 Python API를 쓰려면 ultralytics 설치
!pip install ultralytics

from ultralytics import YOLO
import os
```

```python
# 현재 작업 디렉토리 확인 (필요하면 경로 맞게 수정)
os.getcwd()

# 절대경로 예시 (상대경로도 OK)
DATA_YAML_PATH = os.path.abspath("food_data_3/data.yaml")
print("✅ data.yaml 경로:", DATA_YAML_PATH)
```

```python
# yolov5m.pt (사전학습 가중치)로 시작
model = YOLO('yolov5m.pt')
```

```python
# 프로젝트 폴더: 현재 작업 폴더의 runs 디렉토리
PROJECT_PATH = os.path.abspath("runs")
print("✅ 학습 결과 저장 위치:", PROJECT_PATH)

# ✅ trial_12에서 얻은 best 하이퍼파라미터 조합
best_params = {
    'lr0': 0.000073,
    'momentum': 0.758051,
    'weight_decay': 0.000315,
    'optimizer': 'AdamW'
}

# 학습 실행
model.train(
    data=DATA_YAML_PATH,    # data.yaml 경로
    epochs=100,             # 에포크 수
    imgsz=640,              # 이미지 크기
    batch=16,               # 배치 사이즈
    project=PROJECT_PATH,   # runs/ 폴더에 저장
    name='food_data3_exp_trial12_best',  # 실험 이름 (구분용)
    save_period=1,          # 매 에포크마다 저장
    **best_params           # ⭐ 하이퍼파라미터 주입
)
```
