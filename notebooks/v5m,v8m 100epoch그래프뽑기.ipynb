```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 
df = pd.read_csv("1027.csv")

#   
epoch = df["epoch"]
precision = df["metrics/precision(B)"]

#  precision 
max_idx = precision.idxmax()          #    
max_epoch = epoch[max_idx]            #  epoch
max_precision = precision[max_idx]    #  precision 

print(f"Max precision(B): {max_precision:.4f} at epoch {max_epoch}")

#  
plt.figure(figsize=(14, 6))  #   
plt.plot(epoch, precision,
         linewidth=2,
         marker='o',
         markersize=4,        #   
         markerfacecolor='royalblue',
         markeredgecolor='royalblue',
         color='royalblue')

#  precision   (  +  )
plt.scatter(max_epoch, max_precision, s=60, color='red', zorder=5)
plt.text(
    max_epoch,
    max_precision + 0.01,  #    
    f"max={max_precision:.3f}\nepoch={int(max_epoch)}",
    ha='center',
    va='bottom',
    fontsize=10,
    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="red", lw=1, alpha=0.8)
)

#   /  / 
plt.title("YOLOv5m Training - Epoch vs Precision", fontsize=18)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("Precision (B)", fontsize=14)

plt.grid(True, linestyle='--', alpha=0.5)

#  
plt.tight_layout()

plt.show()
```
