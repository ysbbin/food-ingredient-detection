```python
import os
from ultralytics import YOLO

# ‚úÖ Îç∞Ïù¥ÌÑ∞ÏÖã Í≤ΩÎ°ú
DATA_YAML_PATH = os.path.abspath("food_data_1/data.yaml")
print("‚úîÔ∏è Îç∞Ïù¥ÌÑ∞ YAML Í≤ΩÎ°ú:", DATA_YAML_PATH)

# ‚úÖ Í≤∞Í≥º Ï†ÄÏû• Í≤ΩÎ°ú (Î™®Îç∏Î≥ÑÎ°ú Îã§Î•¥Í≤å ÏßÄÏ†ï)
RESULTS_DIR = "runs_model_compare_0816_1Î≤àÎç∞Ïù¥ÌÑ∞"
os.makedirs(RESULTS_DIR, exist_ok=True)

# ‚úÖ ÌïôÏäµ Í≥µÌÜµ ÏÑ§Ï†ï
EPOCHS = 30
IMG_SIZE = 640
BATCH_SIZE = 16
```

```python
def train_yolo_model(model_name, save_name):
    """
    model_name: 'yolov5s', 'yolov5m', 'yolov8s', 'yolov8m'
    save_name: Í≤∞Í≥º Ï†ÄÏû• Ìè¥ÎçîÎ™Ö
    """
    print(f"\nüöÄ Î™®Îç∏ ÌïôÏäµ ÏãúÏûë: {model_name}")

    model = YOLO(f"{model_name}.pt")  # ÏÇ¨Ï†ÑÌïôÏäµ Í∞ÄÏ§ëÏπò

    results = model.train(
        data=DATA_YAML_PATH,
        epochs=EPOCHS,
        imgsz=IMG_SIZE,
        batch=BATCH_SIZE,
        project=RESULTS_DIR,
        name=save_name,
        save=True,
        device=0,
        verbose=True,
    )

    # ‚úÖ Í≤∞Í≥º Ï†ÄÏû• ÎîîÎ†âÌÜ†Î¶¨
    result_path = os.path.join(RESULTS_DIR, save_name, "results.csv")
    print(f"üìÅ Í≤∞Í≥º CSV Ï†ÄÏû• ÏúÑÏπò: {result_path}")
    return result_path
```

```python
csv_paths = {}

csv_paths["yolov5s"] = train_yolo_model("yolov5s", "exp_yolov5s")
csv_paths["yolov5m"] = train_yolo_model("yolov5m", "exp_yolov5m")
csv_paths["yolov8s"] = train_yolo_model("yolov8s", "exp_yolov8s")
csv_paths["yolov8m"] = train_yolo_model("yolov8m", "exp_yolov8m")
```

```python
import pandas as pd

# ‚úÖ Î™®Îç∏Î≥Ñ results.csv Í≤ΩÎ°ú
csv_paths = {
    "v5s": "runs_model_compare_0816_1Î≤àÎç∞Ïù¥ÌÑ∞/exp_yolov5s/results.csv",
    "v5m": "runs_model_compare_0816_1Î≤àÎç∞Ïù¥ÌÑ∞/exp_yolov5m/results.csv",
    "v8s": "runs_model_compare_0816_1Î≤àÎç∞Ïù¥ÌÑ∞/exp_yolov8s/results.csv",
    "v8m": "runs_model_compare_0816_1Î≤àÎç∞Ïù¥ÌÑ∞/exp_yolov8m/results.csv"
}

# ‚úÖ ÏßÄÌëú Ïª¨Îüº ÌõÑÎ≥¥(ÌôòÍ≤ΩÎßàÎã§ Ìó§ÎçîÍ∞Ä Ï°∞Í∏àÏî© Îã§Î•º Ïàò ÏûàÏñ¥ÏÑú ÏïàÏ†ÑÌïòÍ≤å)
CAND_COLS = {
    "precision": ["metrics/precision(B)", "precision"],
    "recall":    ["metrics/recall(B)", "recall"],
    "mAP50":     ["metrics/mAP50(B)", "mAP50"],
    "mAP5095":   ["metrics/mAP50-95(B)", "mAP50-95", "metrics/mAP50-95"]
}

def find_col(df, candidates):
    """ÌõÑÎ≥¥ Î¶¨Ïä§Ìä∏ Ï§ë dfÏóê Ï°¥Ïû¨ÌïòÎäî Ï≤´ Ïª¨ÎüºÎ™Ö Î∞òÌôò"""
    for c in candidates:
        if c in df.columns:
            return c
    # Î∂ÄÎ∂Ñ Îß§Ïπ≠(ÎßàÏßÄÎßâ ÏïàÏ†ÑÏû•Ïπò)
    low_cols = {c.lower(): c for c in df.columns}
    for c in candidates:
        key = c.lower()
        if key in low_cols:
            return low_cols[key]
        # 'precision' Ï≤òÎüº ÌÇ§ÏõåÎìúÍ∞Ä Ìè¨Ìï®Îêú Í≤ΩÏö∞
        for lc, orig in low_cols.items():
            if key in lc:
                return orig
    raise KeyError(f"Ïª¨ÎüºÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏùå: {candidates}")

def find_epoch_col(df):
    """epoch Ïª¨Îüº(Ïù¥Î¶ÑÏóê Í≥µÎ∞± ÏÑûÏù∏ Í≤ΩÏö∞ Ìè¨Ìï®) ÌÉêÏÉâ"""
    for c in df.columns:
        if "epoch" in c.lower():
            return c
    return None  # ÏóÜÏúºÎ©¥ Ïù∏Îç±Ïä§Î°ú ÎåÄÏ≤¥

# ‚úÖ Í≤∞Í≥º Ï†ÄÏû• Î¶¨Ïä§Ìä∏
summary_rows = []

for model_name, csv_path in csv_paths.items():
    df = pd.read_csv(csv_path)
    df.columns = df.columns.str.strip()  # ÏïûÎí§ Í≥µÎ∞± Ï†úÍ±∞

    # üîé ÏßÄÌëú Ïª¨Îüº Í≤∞Ï†ï
    col_precision = find_col(df, CAND_COLS["precision"])
    col_recall    = find_col(df, CAND_COLS["recall"])
    col_map50     = find_col(df, CAND_COLS["mAP50"])
    col_map5095   = find_col(df, CAND_COLS["mAP5095"])

    # ‚úÖ ÎßàÏßÄÎßâ ÏóêÌè¨ÌÅ¨(ÎßàÏßÄÎßâ Ìñâ) ÏÇ¨Ïö©
    last_row = df.iloc[-1]
    epoch_col = find_epoch_col(df)
    last_epoch = int(last_row[epoch_col]) if epoch_col is not None else int(df.index[-1])

    summary_rows.append({
        "Model": model_name,
        "Last Epoch": last_epoch,
        "Precision": round(float(last_row[col_precision]), 4),
        "Recall": round(float(last_row[col_recall]), 4),
        "mAP50": round(float(last_row[col_map50]), 4),
        "mAP50-95": round(float(last_row[col_map5095]), 4)
    })

# ‚úÖ ÌëúÎ°ú Ï∂úÎ†• (ÎπÑÍµê Ïö©Ïù¥ÌïòÎèÑÎ°ù Precision ÎÇ¥Î¶ºÏ∞®Ïàú Ï†ïÎ†¨ Ïú†ÏßÄ)
summary_df = pd.DataFrame(summary_rows).sort_values(by="Precision", ascending=False)

print("üìä Î™®Îç∏Î≥Ñ 'ÎßàÏßÄÎßâ ÏóêÌè¨ÌÅ¨' Í∏∞Ï§Ä ÏÑ±Îä• ÏöîÏïΩ:")
try:
    display(summary_df)  # Jupyter/Colab
except NameError:
    print(summary_df.to_string(index=False))  # VSCode/ÌÑ∞ÎØ∏ÎÑê ÎåÄÏïà
```

```python
import pandas as pd

# ‚úÖ Î™®Îç∏Î≥Ñ results.csv Í≤ΩÎ°ú
csv_paths = {
    "v5s": "runs_model_compare_0816_2Î≤àÎç∞Ïù¥ÌÑ∞/exp_yolov5s/results.csv",
    "v5m": "runs_model_compare_0816_2Î≤àÎç∞Ïù¥ÌÑ∞/exp_yolov5m/results.csv",
    "v8s": "runs_model_compare_0816_2Î≤àÎç∞Ïù¥ÌÑ∞/exp_yolov8s/results.csv",
    "v8m": "runs_model_compare_0816_2Î≤àÎç∞Ïù¥ÌÑ∞/exp_yolov8m/results.csv"
}

# ‚úÖ ÏßÄÌëú Ïª¨Îüº ÌõÑÎ≥¥(ÌôòÍ≤ΩÎßàÎã§ Ìó§ÎçîÍ∞Ä Ï°∞Í∏àÏî© Îã§Î•º Ïàò ÏûàÏñ¥ÏÑú ÏïàÏ†ÑÌïòÍ≤å)
CAND_COLS = {
    "precision": ["metrics/precision(B)", "precision"],
    "recall":    ["metrics/recall(B)", "recall"],
    "mAP50":     ["metrics/mAP50(B)", "mAP50"],
    "mAP5095":   ["metrics/mAP50-95(B)", "mAP50-95", "metrics/mAP50-95"]
}

def find_col(df, candidates):
    """ÌõÑÎ≥¥ Î¶¨Ïä§Ìä∏ Ï§ë dfÏóê Ï°¥Ïû¨ÌïòÎäî Ï≤´ Ïª¨ÎüºÎ™Ö Î∞òÌôò"""
    for c in candidates:
        if c in df.columns:
            return c
    # Î∂ÄÎ∂Ñ Îß§Ïπ≠(ÎßàÏßÄÎßâ ÏïàÏ†ÑÏû•Ïπò)
    low_cols = {c.lower(): c for c in df.columns}
    for c in candidates:
        key = c.lower()
        if key in low_cols:
            return low_cols[key]
        # 'precision' Ï≤òÎüº ÌÇ§ÏõåÎìúÍ∞Ä Ìè¨Ìï®Îêú Í≤ΩÏö∞
        for lc, orig in low_cols.items():
            if key in lc:
                return orig
    raise KeyError(f"Ïª¨ÎüºÏùÑ Ï∞æÏùÑ Ïàò ÏóÜÏùå: {candidates}")

def find_epoch_col(df):
    """epoch Ïª¨Îüº(Ïù¥Î¶ÑÏóê Í≥µÎ∞± ÏÑûÏù∏ Í≤ΩÏö∞ Ìè¨Ìï®) ÌÉêÏÉâ"""
    for c in df.columns:
        if "epoch" in c.lower():
            return c
    return None  # ÏóÜÏúºÎ©¥ Ïù∏Îç±Ïä§Î°ú ÎåÄÏ≤¥

# ‚úÖ Í≤∞Í≥º Ï†ÄÏû• Î¶¨Ïä§Ìä∏
summary_rows = []

for model_name, csv_path in csv_paths.items():
    df = pd.read_csv(csv_path)
    df.columns = df.columns.str.strip()  # ÏïûÎí§ Í≥µÎ∞± Ï†úÍ±∞

    # üîé ÏßÄÌëú Ïª¨Îüº Í≤∞Ï†ï
    col_precision = find_col(df, CAND_COLS["precision"])
    col_recall    = find_col(df, CAND_COLS["recall"])
    col_map50     = find_col(df, CAND_COLS["mAP50"])
    col_map5095   = find_col(df, CAND_COLS["mAP5095"])

    # ‚úÖ ÎßàÏßÄÎßâ ÏóêÌè¨ÌÅ¨(ÎßàÏßÄÎßâ Ìñâ) ÏÇ¨Ïö©
    last_row = df.iloc[-1]
    epoch_col = find_epoch_col(df)
    last_epoch = int(last_row[epoch_col]) if epoch_col is not None else int(df.index[-1])

    summary_rows.append({
        "Model": model_name,
        "Last Epoch": last_epoch,
        "Precision": round(float(last_row[col_precision]), 4),
        "Recall": round(float(last_row[col_recall]), 4),
        "mAP50": round(float(last_row[col_map50]), 4),
        "mAP50-95": round(float(last_row[col_map5095]), 4)
    })

# ‚úÖ ÌëúÎ°ú Ï∂úÎ†• (ÎπÑÍµê Ïö©Ïù¥ÌïòÎèÑÎ°ù Precision ÎÇ¥Î¶ºÏ∞®Ïàú Ï†ïÎ†¨ Ïú†ÏßÄ)
summary_df = pd.DataFrame(summary_rows).sort_values(by="Precision", ascending=False)

print("üìä Î™®Îç∏Î≥Ñ 'ÎßàÏßÄÎßâ ÏóêÌè¨ÌÅ¨' Í∏∞Ï§Ä ÏÑ±Îä• ÏöîÏïΩ:")
try:
    display(summary_df)  # Jupyter/Colab
except NameError:
    print(summary_df.to_string(index=False))  # VSCode/ÌÑ∞ÎØ∏ÎÑê ÎåÄÏïà
```
